```python
import pandas as pd
import numpy as np
import requests
```

```python
url = "https://www.starbucks.co.kr/store/getStore.do?r=RO98U559C0"

# 시,도 분류코드를 star_list로 담고
star_list = ['01','02','03','04','05','06','07','08','09',
             '10','11','12','13','14','15','16','17']
# 01 서울, 02 광주, 03 대구, 04 대전, 05 부산, 06 울산, 07 인천, 08 경기, 09 강원,
# 10 경남, 11 경북, 12 전남, 13 전북, 14 충남, 15 충북, 16 제주, 17 세종
# 각 시도의 데이터를 json타입으로 데이터를 받을 star_json 리스트 생성.
star_json = []

# 각 시도 순서대로 request하여 json타입의 데이터를 star_json에 넣는다
for i in star_list:
    response = requests.post(url,data={
    'p_sido_cd' : i,                    # 시,도 분류 코드
    'p_gugun_cd': '',                   # 시,군,구 분류 코드
    'in_biz_cd' : '' ,                  # 알 수 없는 정보
    'iend'      : 2000,                 # 불러올 최대 매장 수
    'set_date'  : ''                    # 알 수 없는 정보
    })
    star_json.append(response.json())
```

```python
# 각 시도별 json타입으로 만든 데이터를 순차적으로 데이터프레임으로 만들고
# concat으로 생성 즉시 붙여서 df 변수에 넣는다
df = pd.concat(pd.DataFrame(star_json[i]['list']) for i in range(17))
```

```python
# df 확인
pd.set_option('display.max_columns', None)
print(df.shape)
df.head()
```

```python
# 필요한 컬럼만 부르고, 한글명 컬럼으로 변환한다.
col = ['지점명', '시도', '시군구', '도로명주소', '위도', '경도']
star_df = df[['s_name', 'sido_name', 'gugun_name'
                       , 'doro_address', 'lat', 'lot']]
star_df.columns = col
# 인덱스 초기화
star_df.reset_index(drop=True, inplace=True)
star_df
```

```python
# 시도별 지점 수 확인
star_df['시도'].value_counts()
```

```python
# '강원도'로 표기된 데이터 '강원'으로 변경
star_df['시도'] = star_df['시도'].apply(lambda x : '강원' if x == '강원도' else x)
star_df['시도'].value_counts()
```
